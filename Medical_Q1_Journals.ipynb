## Libraries and Data

```python
import pandas as pd
from statistics import mean, median
import matplotlib.pyplot as plt

df_full = pd.read_csv('scimagojr_medicine_2023.csv', sep = ';', decimal = ',') # load medical journal data

df_full # display first and last 5 rows of dataframe
```

## Medical

We are interested in ***Q1 journals***, also known as the top quartile (top 25%) of medical journals based on the SJR ranking. We'll also only filter journals that had ***at least one publication*** in 2023.

```python
df = df_full[df_full['SJR Best Quartile'] == 'Q1'].reset_index() # Only Q1 journals
df = df[df['Total Docs. (2023)'] > 1] # At least one paper published in 2023

df # display first and last 5 rows of dataframe
```

Of all medical journals, 1,928 are considered Q1 journals. Let's answer some exploratory and descriptive questions based on the available data.

***How many papers were published across all Q1 medical journals in 2023?***

```python
total_medical = sum(df['Total Docs. (2023)'])

print(f'In 2023 there were {total_medical:,} papers published in Q1 medical journals.')
```

***What were the top 5 highest ranked Q1 medical journals in 2023?***

```python
df[['Rank','Title','SJR','H index','Total Docs. (2023)','%Female','Country']][:5]
```

***What were the journals with most and least papers published in 2023?***

```python
df.sort_values(by = 'Total Docs. (2023)', ascending = False)
```

```python
# store sorted df as a new variable
medical_sorted = df.sort_values(by = 'Total Docs. (2023)', ascending = False)

# Q1 medical journal with most papers published in 2023
medical_most = medical_sorted.iloc[0]['Title'] # journal name
medical_most_n = medical_sorted.iloc[0]['Total Docs. (2023)'] # papers published

# Q1 medical journal with least papers published in 2023
medical_least = medical_sorted.iloc[-1]['Title'] # journal name
medical_least_n = medical_sorted.iloc[-1]['Total Docs. (2023)'] # papers published

print(f'"{medical_most}" had the most papers published in 2023 with {medical_most_n:,} publications.')
print(f'"{medical_least}" had the most papers published in 2023 with {medical_least_n:,} publications.')
```

***What was the median amount of papers published in Q1 medical journals in 2023?***

```python
medical_median_n = median(df['Total Docs. (2023)'])

print(f'In 2023, the median amount of papers published in Q1 medical journals was {int(medical_median)}.')
```

***What was the average percentage of female authors in Q1 medical journals in 2023?***

```python
medical_mean_f = round(mean(df['%Female']), 1)

print(f'The mean percentage of female authors in Q1 medical journals in 2023 was {medical_mean_f}%.')
```

***What are the top 5 countries with most Q1 medical Journals and how many journals do they account for?***

```python
# Top 5 countries with most Q1 medical journal
medical_top5 = df.groupby(['Country']).size().to_frame('Count').sort_values('Count', ascending = False).reset_index()[:5]

# Add column indicating percentage of total
medical_top5['%'] = round(medical_top5['Count'] / len(df.index) * 100, 1)

medical_top5
```
